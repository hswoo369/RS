```python
import gopax_api_wrapper as gopax
import numpy as np
import pandas as pd
import matplotlib
from datetime import datetime
```

```python
#data frame을 받아서 표로 보여줌
def get_past_dataframe(trading_pair, start, end, interval):
    start_ts = gopax.get_timestamp(start)
    end_ts = gopax.get_timestamp(end)
    raw_result = gopax.get_past_data(trading_pair, start_ts, end_ts, interval)
    result = pd.DataFrame(raw_result, columns = ['time', 'low', 'high', 'open', 'close', 'volume'])
    return result
```

```python
diff = 9 * (3600)
```

```python
result = get_past_dataframe('BTC-KRW', '2018-04-15 00:00', '2018-04-17 00:00', 30)
result['time'] = result['time'].apply(lambda x: datetime.utcfromtimestamp(x // 1000))
```

```python
def get_clean_data(trading_pair, start, end, interval):
    diff = 9 * 3600
    s = start
    e = end
    ts = gopax.get_timestamp(s)
    te = gopax.get_timestamp(e)
    result = get_past_dataframe(trading_pair, s, e, interval)
    result = result[result.time >= ts * 1000]
    result = result[result.time < te * 1000]
    result['time'] = result['time'].apply(lambda x: datetime.utcfromtimestamp((x // 1000) + diff))
    return result
    
```

```python
df = get_clean_data('BTC-KRW', '2018-03-05 00:00', '2018-03-20 00:00', 30)
df = df.set_index('time')
df['close'].plot(figsize = (15,5))
```

```python
# 초기 추세선 설정
# 현재 step에서 4개 이전과의 slope를 구했을 때
# slope가 양이면 초기 추세선으로 설정
# slope가 음이면 다음 step으로 넘어가도록 break

def init_trend_line(close_value, step):
    slope = -1000000000000000000
    while slope <= 0 :
        slope = (close_value[step] - close_value[step-4])/4
        if slope > 0:
            trend_line = slope
            return trend_line
        else:
            break
```

```python
init_trend_line(close_value, 16)
```

```python
# 상승 추세선 판단
# init_trend_line과 4개 이후의 데이터간 slope와 비교
# slope가 trendline보다 두번연속 증가하면 증가추세라고 판단
# trendline을 최신화
#slope가 두번연속 증가하지 않는다면 

def increase_trend_line(close_value, trend_line, step):
    if (trend_line > 0) & \
    ((close_value[step-2] - close_value[step-8])/6 > 0) & \
    ((close_value[step] - close_value[step-8])/8 > 0):
        slope = (close_value[step] - close_value[step-8])/8
        return slope
```
